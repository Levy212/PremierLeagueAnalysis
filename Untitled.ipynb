```python
import pandas as pd
import numpy as np

df = pd.read_csv('PremierLeague.csv')
df.columns
```

```python
df.head()
```

```python
#classement du nombre de victoire depuis 1993
home_wins = df[df['FullTimeResult'] == 'H'].groupby('HomeTeam').size()
away_wins = df[df['FullTimeResult'] == 'A'].groupby('AwayTeam').size()
total_wins = home_wins.add(away_wins, fill_value=0).sort_values(ascending=False)
total_wins.reset_index().style.background_gradient(cmap="CMRmap_r")
```
