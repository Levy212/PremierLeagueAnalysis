```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv('PremierLeague.csv')
df.columns
```

```python
df.head()
```

```python
#classement du nombre de victoire depuis 1993
home_wins = df[df['FullTimeResult'] == 'H'].groupby('HomeTeam').size()

away_wins = df[df['FullTimeResult'] == 'A'].groupby('AwayTeam').size()
total_wins = home_wins.add(away_wins, fill_value=0).sort_values(ascending=False)
total_wins.reset_index().style.background_gradient(cmap="CMRmap_r")
```

```python
df['TotalYellowCards'] = df['HomeTeamYellowCards'] + df['AwayTeamYellowCards']
df['TotalRedCards'] = df['HomeTeamRedCards'] + df['AwayTeamRedCards']

referee_cards = df.groupby('Referee').agg(
    {'TotalYellowCards': 'sum', 
     'TotalRedCards': 'sum', 
     'Date': 'max', 
     'MatchID': 'count'}
).sort_values(by='TotalYellowCards', ascending=False)

# Renommer la colonne 'MatchID' pour indiquer qu'il s'agit du nombre total de matchs
referee_cards = referee_cards.rename(columns={'MatchID': 'TotalMatches'})


referee_cards.head()
```

```python
top_referees_cards = referee_cards.head(10)

# Calculer les ratios pour les cartons jaunes et rouges par match pour chaque arbitre
top_referees_cards['YellowCardsPerMatch'] = top_referees_cards['TotalYellowCards'] / top_referees_cards['TotalMatches']
top_referees_cards['RedCardsPerMatch'] = top_referees_cards['TotalRedCards'] / top_referees_cards['TotalMatches']
top_referees_cards.head()
```

```python
season_cards = df.groupby('Season').agg(
    {'TotalYellowCards': 'sum', 
     'TotalRedCards': 'sum',
    'MatchID': 'count'}).sort_values(by='Season', ascending=True)
season_cards = season_cards[season_cards['TotalRedCards']>7]
print(season_cards)
```

```python
#focus on M Dean

dean_matches = df[df['Referee']=="M Dean"]
dean_matches.describe()
```

```python
dean_matches.head()
```

```python
home_team_yellow_cards = dean_matches.groupby("HomeTeam")["HomeTeamYellowCards"].sum()
away_team_yellow_cards = dean_matches.groupby("AwayTeam")["AwayTeamYellowCards"].sum()
total_yellow_cards = home_team_yellow_cards.add(away_team_yellow_cards, fill_value=0).sort_values(ascending=False)
total_yellow_cards.head()
```

```python
home_team_matches = dean_matches.groupby("HomeTeam")["MatchID"].count()
away_team_matches = dean_matches.groupby("AwayTeam")["MatchID"].count()
total_matches = home_team_matches.add(away_team_matches, fill_value=0)
yellow_cards_per_match = total_yellow_cards / total_matches
dean_team_stats = pd.DataFrame({
    "TotalYellowCards": total_yellow_cards,
    "TotalMatches": total_matches,
    "YellowCardsPerMatch": yellow_cards_per_match
}).sort_values(by="YellowCardsPerMatch", ascending=False)
dean_team_stats = dean_team_stats[dean_team_stats["TotalMatches"]>10]
dean_team_stats.describe()
```
